```python
def convertir_hora_a_periodo(hora):
    if 0 <= hora <= 5:
        return 'Madrugada'
    elif 6 <= hora <= 11:
        return 'Mañana'
    elif hora == 12:
        return 'Medio día'
    elif 13 <= hora <= 17:
        return 'Tarde'
    elif 18 <= hora <= 23:
        return 'Noche'
    else:
        return 'Desconocido'  # Por si hay valores fuera de rango

data_train['horas'] = data_train['horas'].apply(convertir_hora_a_periodo)
```

# Edad

```python
# Vamos a crear una variable edad, calculando en base la fecha de la transacción y la fecha de nacimiento.

data_train['dob'] = pd.to_datetime(data_train['dob'], format='%Y-%m-%d')

delta = data_train['trans_date_trans_time'] - data_train['dob']

data_train['edad'] = delta.dt.days // 365 # Dividimos entre 365 para obtener la edad en años
data_train['edad'].min() # Vamos a ver la edad de los clientes.
```

```python
# Vamos a crear grupos de edad con cut 

bins = [0, 18, 25, 35, 45, 55, 65, 100]
labels = ['0-18', '19-25', '26-35', '36-45', '46-55', '56-65', '66+']
data_train['grupo_edad'] = pd.cut(data_train['edad'], bins=bins, labels=labels, right=False) 
data_train['grupo_edad'].value_counts()
```

# Distancia 

```python
# Calcular la distancia entre las coordenadas del negocio y del cliente

from math import radians, cos, sin, asin, sqrt

def haversine(lat1, lon1, lat2, lon2):
    # Radio de la Tierra en kilómetros
    R = 6371.0

    # Convertir grados a radianes
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

    # Diferencias
    dlat = lat2 - lat1
    dlon = lon2 - lon1

    # Fórmula de Haversine
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * asin(sqrt(a))

    # Distancia
    distance = R * c
    return distance

```

```python
data_train['distancia_cliente_negocio'] = data_train.loc[:,['lat', 'long', 'merch_lat', 'merch_long']].apply(lambda x: haversine(x['lat'], x['long'], x['merch_lat'], x['merch_long']), axis=1)
```

# Identificamos tarjetas de créditos

```python
def card_type(cc_num):
    if (len(cc_num) == 16 or len (cc_num) == 13 or len (cc_num) == 19) and cc_num.startswith('4'):
        return 'Visa'
    elif len(cc_num) == 16 and (cc_num.startswith('5155') or cc_num.startswith('2221') or cc_num.startswith('2720')):
        return 'MasterCard'
    elif len(cc_num) == 15 and (cc_num.startswith('34') or cc_num.startswith('37')):
        return 'American Express'
    elif len(cc_num) == 16 and (cc_num.startswith('6011') or cc_num.startswith('622') or cc_num.startswith('64') or cc_num.startswith('65')):
        return 'Discover'
    elif (len(cc_num) == 16 or len(cc_num) == 14) and (cc_num.startswith('300') or cc_num.startswith('305') or cc_num.startswith('36') or cc_num.startswith('38') or cc_num.startswith('39')):
        return 'Dinner Club'
    else :
        return 'Unknown'
```

```python
# Aplicamos la función a la columna 'cc_num' y creamos una nueva columna 'card_type'
data_train['card_type'] = data_train['cc_num'].astype('str').apply(card_type)
data_train['card_type'].value_counts()
```

# Creamos el tipo de ubicación

```python
tipos_via = ['Avenue', 'Street', 'Boulevard', 'Drive', 'Road', 'Lane', 'Court', 'Place', 'Parkway', 'Circle', 'Trail', 'Way', 'Terrace', 'Loop', 'Creek', 'Brook', 'Manor', 'Ridge', 'View']

import re
def extraer_tipo_via(direccion):
    for tipo in tipos_via:
        if re.search(rf'\b{tipo}\b', direccion, re.IGNORECASE):
            return tipo
    return 'Otro'

data_train['tipo_via'] = data_train['street'].apply(extraer_tipo_via)
```

# Identificamos empleos 

```python
def agrupar_ocupacion(ocupacion):
    if "nurse" in ocupacion.lower() or "therapist" in ocupacion.lower() or "psychologist" in ocupacion.lower() or "doctor" in ocupacion.lower():
        return "Salud"
    elif "teacher" in ocupacion.lower() or "lecturer" in ocupacion.lower():
        return "Educación"
    elif "engineer" in ocupacion.lower() or "developer" in ocupacion.lower():
        return "Ingeniería"
    elif "scientist" in ocupacion.lower() or "chemist" in ocupacion.lower():
        return "Ciencias"
    elif "accountant" in ocupacion.lower() or "banker" in ocupacion.lower() or "finance" in ocupacion.lower():
        return "Finanzas"
    elif "artist" in ocupacion.lower() or "writer" in ocupacion.lower() or "musician" in ocupacion.lower():
        return "Artes"
    elif "manager" in ocupacion.lower() or "consultant" in ocupacion.lower() or "executive" in ocupacion.lower():
        return "Negocios"
    elif "architect" in ocupacion.lower() or "survey" in ocupacion.lower():
        return "Construcción"
    elif "officer" in ocupacion.lower() or "administrator" in ocupacion.lower():
        return "Sector público"
    else:
        return "Otros"

data_train['job'] = data_train['job'].apply(agrupar_ocupacion)
```
